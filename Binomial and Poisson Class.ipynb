```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
n=100
```

```python
p=.1
```

```python
k=np.arange(0,100)
```

```python
binomial=stats.binom.cdf(k,n,p)
```

```python
binomial
```

```python
plt.plot(k,binomial,'o-')
```

```python
binomial1=stats.binom.pmf(k,n,p)
binomial1
```

```python
plt.plot(k,binomial1,'o-')
```
